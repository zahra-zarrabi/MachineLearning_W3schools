```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
from sklearn.metrics import r2_score  #and  from sklearn import linear_model   
from sklearn.preprocessing import StandardScaler
import pandas
```

```python
# Mean
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
average=np.mean(speed)
print('average :',average)

# Median
median=np.median(speed)
print('Median :',median)

# Mode
mode=stats.mode(speed)
print('Mode :',mode)
```

```python
# Standard Deviation
speed = [86,87,88,86,87,85,86]
standard=np.std(speed)
print('Standard Deviation :',standard)

# Variance
speed = [32,111,138,28,59,77,97]
variance=np.var(speed)
print('Variance:',variance)
```

```python
# Percentiles
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
percentiles=np.percentile(ages,80)
print('Percentiles :',percentiles)
```

```python
# Linear Regression
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
#Train/Test
np.random.seed(2)

x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = np.poly1d(np.polyfit(train_x, train_y, 4))

myline = np.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()

r2 = r2_score(test_y, mymodel(test_x))
print(r2)
```
